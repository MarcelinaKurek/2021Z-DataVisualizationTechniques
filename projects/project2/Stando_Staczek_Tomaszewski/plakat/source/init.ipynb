### Lollipop

```r
# 2020

suppressMessages(df_plik <- read_excel("data/deaths/2020.xlsx", 1))
df_plik <- df_plik[6:length(df_plik$`...10`), ]
df_plik[1, 1] <- "Wiek"
df_plik[1, 2] <- "Region_kod"
df_plik[1, 3] <- "wojewodztwo"

wojewodztwa <- c("Dolnośląskie", "Kujawsko-Pomorskie", "Lubelskie", "Lubuskie", "Łódzkie", "Małopolskie", "Makroregion Województwo Mazowieckie", "Opolskie", "Podkarpackie", "Podlaskie", "Pomorskie", "Śląskie", "Świętokrzyskie", "Warmińsko-Mazurskie", "Wielkopolskie", "Zachodniopomorskie")

colnames(df_plik) <- as.character(as.vector(df_plik[1, ]))
df_plik <- df_plik[4:length(df_plik$T01), ]

df_plik <- df_plik %>%
    filter(Wiek == "Ogółem") %>%
    select(wojewodztwo, T40, T41, T42, T43, T44) %>%
    filter(wojewodztwo %in% wojewodztwa) %>%
    mutate(T40 = as.numeric(T40)) %>%
    mutate(T41 = as.numeric(T41)) %>%
    mutate(T42 = as.numeric(T42)) %>%
    mutate(T43 = as.numeric(T43)) %>%
    mutate(T44 = as.numeric(T44)) %>%
    mutate(rok_2020 = T40 + T41 + T42 + T43 + T44) %>%
    select(wojewodztwo, rok_2020)

df_plik$wojewodztwo <- ifelse(df_plik$wojewodztwo == "Makroregion Województwo Mazowieckie", "Mazowieckie", df_plik$wojewodztwo)

df <- df_plik # koncowa tabela

df_plik
```

```r
# 2019

suppressMessages(df_plik <- read_excel("data/deaths/2019.xlsx", 1))
df_plik <- df_plik[6:length(df_plik$`...10`), ]
df_plik[1, 1] <- "Wiek"
df_plik[1, 2] <- "Region_kod"
df_plik[1, 3] <- "wojewodztwo"

wojewodztwa <- c("Dolnośląskie", "Kujawsko-Pomorskie", "Lubelskie", "Lubuskie", "Łódzkie", "Małopolskie", "Makroregion Województwo Mazowieckie", "Opolskie", "Podkarpackie", "Podlaskie", "Pomorskie", "Śląskie", "Świętokrzyskie", "Warmińsko-Mazurskie", "Wielkopolskie", "Zachodniopomorskie")

colnames(df_plik) <- as.character(as.vector(df_plik[1, ]))
df_plik <- df_plik[4:length(df_plik$T01), ]
df_plik <- df_plik %>%
    filter(Wiek == "Ogółem") %>%
    select(wojewodztwo, T40, T41, T42, T43, T44) %>%
    filter(wojewodztwo %in% wojewodztwa) %>%
    mutate(T40 = as.numeric(T40)) %>%
    mutate(T41 = as.numeric(T41)) %>%
    mutate(T42 = as.numeric(T42)) %>%
    mutate(T43 = as.numeric(T43)) %>%
    mutate(T44 = as.numeric(T44)) %>%
    mutate(rok_2019 = T40 + T41 + T42 + T43 + T44) %>%
    select(wojewodztwo, rok_2019)

df_plik$wojewodztwo <- ifelse(df_plik$wojewodztwo == "Makroregion Województwo Mazowieckie", "Mazowieckie", df_plik$wojewodztwo)

df <- full_join(df, df_plik, by = "wojewodztwo") # koncowa tabela
df
```

```r
df_plik <- read.csv("data/covid_woj.csv", sep=",", stringsAsFactors = FALSE)
df_plik <- df_plik[69:86, ]
colnames(df_plik) <- as.character(as.vector(df_plik[2, ]))
df_plik <- df_plik[3:length(df_plik$`Województwo`), ]

df_plik[ , 3:ncol(df_plik)] <- sapply(df_plik[ , 3:ncol(df_plik)], as.numeric)

df_plik$rok_2020_covid_zgony <- df_plik$'01.11' - df_plik$'27.09'

df_plik <- df_plik %>% select(Województwo, rok_2020_covid_zgony)

colnames(df_plik) <- c("wojewodztwo", "rok_2020_covid_zgony")

df <- full_join(df, df_plik, by = "wojewodztwo") # koncowa tabela
df
```
