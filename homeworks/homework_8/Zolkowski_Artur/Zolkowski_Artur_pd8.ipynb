## Poprawa w Pythonie

```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
sns.set(rc={'figure.figsize':(12,7)})
sns.set_style("darkgrid")
pd.options.display.max_rows=10
pd.set_option('display.width', 5000) 
pd.set_option('display.max_columns', 60)
sns.set_theme(style="whitegrid")
```

```python
data = {"Wielka Brytania":626246,"Izrael":609406,
        "USA":579482,"Belgia":507841,  "Hiszpania":468695,
        "Portugalia":427248, "Australia":390289,
        "Włochy":361021, "Austria":343202,
        "Niemcy":332072, "Szwajcaria":314967,
        "Francja":313959, "Szwecja":313550,
        "Czechy":285370, "Holandia":235763,
        "Rumunia":214115, "Słowacja":196763,
        "Polska":164872
        }
s = pd.Series(data)
```

```python
ax = sns.barplot(x=s.index, y=s.values,
           color = "grey")
plt.xticks(rotation=20, ha="right")
plt.title("Liczba testów na 1mln populacji w wybranych krajach", fontsize=20)
vals = ax.get_xticks()

plt.show()
```
