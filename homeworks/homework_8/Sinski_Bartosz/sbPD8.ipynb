## Poprawiona Wizualizacja


```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
import matplotlib.ticker as tic

zboza = ['Hard red\nwinter wheat','Corn','Soybean','Soft red\nwinter wheat']
wzrost = [32.9,31.8,31.5,20.7]
df =  pd.DataFrame({'Crops':zboza, 'Percent Growth':wzrost})
plt.figure(figsize=(12,8))
sns.set_style("whitegrid")
a1 = sns.barplot(data=df,x = "Crops",y="Percent Growth", color = "dimgrey")
plt.ylim(0,35)
a1.yaxis.set_major_formatter(tic.PercentFormatter())
a1.set(xlabel = None,ylabel =None)
plt.title("Percent growth in agricultural futures price over last three months, by crop",loc="left")
sns.despine(left=True, bottom=True)
```

## Uzasadnienie 
Użycie grafik zbóż jako słupków w oryginalnej wizualizacji bardzo negatywnie wpływa na czytelność wykresu.
