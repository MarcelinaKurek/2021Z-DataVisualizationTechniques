## Wykres Poprawiony

```python
labels = ['A', 'B', 'C', 'D', "E"]
sizes = [36, 45, 42, 40, 35]
fig, ax = plt.subplots(figsize=(12, 10))
ax.grid(axis="y", alpha=1)
ax.set_title("The girls percentage out of total students who went to \n five different places i.e. A, B, C, D & E to spend their summer vacation.", fontsize=15)
ax.bar(labels, sizes, width = 0.6)
ax.set_ylabel("%", fontsize=15)


rects = ax.patches
for rect, label in zip(rects, sizes):
    height = rect.get_height()
    ax.text(rect.get_x() + rect.get_width() / 2, height + 0.002, label,
    ha='center', va='bottom', fontsize=14)
plt.show()
```

Głownym problmem w wykresie oryginalnym był fakt, że skala nie zaczynała sie od 0. W związku osoba patrząca na wykres może odnieść wrażenie, że do miejsca "B" pojechało ponad 2 razy więcej osób niż do miejsca "A". W rzeczywistości jest to dużo mniejsza rożnica.
