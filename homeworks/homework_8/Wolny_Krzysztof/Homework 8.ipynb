## Mój wykres

```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
import numpy as np
import matplotlib.ticker as mtick
```

```python
# Dane: 
names = ["E-mail", "Searching for goods and services", "Reading news or magazines", "Social network", 
         "Internet banking", "Posting on your website", "Watching TV on-demand", "Selling goods and services"]

values = [71.2, 66.8, 64.8, 55.0, 52.8, 27.2, 23.1, 16.8]

df = pd.DataFrame({"names":names, "value":values})

df = df.sort_values(by='value')
my_range=range(1,len(df.names)+1)
```

```python

fig = plt.figure(figsize=(15,9))


ax = plt.subplot(1, 1, 1)
ax.grid(color='grey', linestyle='-', 
         linewidth=0.5, 
         axis = 'x', 
        )
ax.set_axisbelow(True)
ax.xaxis.set_major_formatter(mtick.PercentFormatter())

ax.hlines(y=my_range, xmin=0, xmax=df.value, color = '#338aff', linewidth = 5)
ax.plot(df.value, my_range, "D", markersize=14)

plt.yticks(my_range, df.names, size = 18)
plt.xticks(size = 15)
plt.xlim(0, 80)

plt.title('Purposes on using the Internet for private use by persons \naged 16–74 in 2019', 
            loc = 'left', 
            fontsize = 25, 
         pad = 10)
plt.show()
```
