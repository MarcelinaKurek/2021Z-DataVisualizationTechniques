Dzięki umieszczeniu legendy na wykresie możemy odrazu jasno przeczytać czego dotyczą poszczególne słupki

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

PIS = [31,59,6,4]
KO = [85,12,1,2]
Polska2050 = [68,27,0,5]
Lewica = [77,21,1,0]
Konfederacja = [17,51,25,7]
PSL = [48,42,0,10]
data = [PIS, KO, Polska2050, Lewica, Konfederacja, PSL]
colNames = ["PIS","KO","Polska2050","Lewica","Konfederacja","PSL"]
X = pd.DataFrame(data = data, index = colNames)


labels = ['PIS', 'KO', 'Polska 2050', 'Lewica', 'Konfederacja', "PSL"]
scislejsza_wspolpraca = [31, 85, 68, 77, 17, 48]
ograniczenie_wspolpracy = [59,12, 27, 21, 51, 42]
wyjscie_z_UE = [6, 1, 0, 1, 25, 0]
nie_wiem = [4, 2, 5, 0, 7, 10]

x = np.arange(len(labels))  # the label locations
width = 0.20  # the width of the bars

fig, ax = plt.subplots()
plt.figure(figsize=(1000, 1000))
rects1 = ax.bar(x - width/2, scislejsza_wspolpraca, width, label='ścisła wspólpraca')
rects2 = ax.bar(x + width/2, ograniczenie_wspolpracy, width, label='ograniczenie wspolpracy')
rects3 = ax.bar(x + 3/2*width, wyjscie_z_UE, width, label='wyjscie z UE')
rects4 = ax.bar(x + 5/2*width, nie_wiem, width, label='Nie wiem')

# Add some text for labels, title and custom x-axis tick labels, etc.

ax.set_ylabel('Wartosc procentowa')
ax.set_title('Jaka polityka według partii')
ax.set_xlabel('Partia')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.grid(axis = "y")
ax.legend(prop={'size': 8})
def autolabel(rects):
    """Attach a text label above each bar in *rects*, displaying its height."""
    for rect in rects:
        height = rect.get_height()
        ax.annotate('{}'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')


autolabel(rects1)
autolabel(rects2)
autolabel(rects3)
autolabel(rects4)

plt.show()

```
