# Poprawiona wizualizacja w Pythonie

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
#import seaborn as sns
```

```python
df = pd.read_csv('data.csv', sep = ',')
df = df.sort_values(by=['2019'], ascending=False)
temp = df.iloc[9].copy()
df = df.drop(df.index[9])
df = df.append(temp, ignore_index=True)
```

```python
plt.figure(figsize=(15, 10))
plt.bar(df["GEO"], df["2019"], color = 'crimson')
plt.yticks(np.arange(0, 25, 1))
plt.xlabel('EU countries', size = 20)
plt.ylabel('Share of population (%)', size = 20)
plt.title('Share of population not having indoor flushing toilet\n for the solo use of their household in 2019', size = 22)
plt.grid(axis='y')
plt.xticks(rotation = 40, ha="right")
for i, v in enumerate(df["2019"]):
    plt.text(i-0.4, v+0.2, str(v), color='darkred', fontweight='bold', size=9)
plt.show()
```

## Uzasadnienie
Oryginalny wykres jest wyjątkowo beznadziejny. Przypomina bardziej grafikę z podpisami państw wstawionymi w losowych miejscach. Ciężko jest go użyć do porównywania bądź nawet odczytywania wartości. Nie wiadomo, co dokładnie ma ukazywać rozwinięta rolka papieru ani w jaki sposób podpisane wartości są z nią powiązane. W części wykresu bliskiej zera ciężko też jest przypisać podpisy państw do zaznaczonych wartości. W teorii mają to sugerować kropki ale nie jest to czytelne.\
Dlatego postanowiłem zastąpić tę wizualizację wykresem słupkowym. Bez problemu można odczytywać i porównywać wartości. Słupki też w sposób czytelny ukazują rząd wielkości wartości. Wykres jest bardziej intuicyjny. Dodatkowo postanowiłem też przenieść słupek EU na bok, aby go wyróżnić.\
Wykres orygnalny ma dosyć niefortunną kolorystykę. Biały na różowym ma bardzo mały kontrast i sprawia, że wykres wydaje się "wyblakły". Pomarańczowe oznaczenia są też mało widoczne. Postanowiłem na mojej wizualizacji dać kolor, który nawiązuje do oryginalnej ale jednak jest dużo bardziej widoczny.\
Dla małych wartości słupki są nieco nieczytelne ale wykres ma przede wszystkim pokazywać rząd wielkości i skrajne różnice panujące wewnątrz Europy.
