### Wykres oryginalny:
![](zly_wykres.jpg "wykres oryginalny")

Wykres znalazłem na grupie na facebooku, został on umieszczony tam wczoraj (03.12.20.). Pochodzi on z artykułu, który można znaleźć pod tym [linkiem](https://www.weforum.org/agenda/2020/11/black-lives-matter-where-are-we-now/).

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
race = ["black", "hispanic", "white", "other"]
deaths_pm = [31, 23, 13, 4]
total_deaths = [1301, 907, 2499, 220]
width = 0.5

fig, (ax1, ax2) = plt.subplots(1, 2)
fig.suptitle('Number of people killed in police shootings in the U.S. since January 1, 2015.', )
fig.subplots_adjust(wspace = 0.3)
fig.set_figwidth(15)
ax1.bar(race, deaths_pm, width, label='deaths_pm', color = "r")
ax1.set_ylabel('deaths per million')
ax2.bar(race, total_deaths, width, label='total_deaths', color = "orange")
ax2.set_ylabel('total deaths')
plt.show()
```

W moim wykresie wyraźnie widać stosunki między danymi rasami niezależnie czy mowa o śmierciach na milion czy o sumie śmierci. W wykresie oryginalnym ciężko porównać do siebie wartości sumy śmierci, ponieważ pola powierzchni kół są trudne do porówanania.

Teraz bez problemu widać, że pomimo tego, że policja w USA zabiła około dwa razy mniej czarnych osób niż białych, współczynnik zabitych czarnych na milion osób jest ponad dwa razy większy niż współczynnik białych.
