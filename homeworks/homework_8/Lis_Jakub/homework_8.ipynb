## Problem z oryginalnym wykresem

Tak naprawdę bardzo cieżko rozczytać ten wykres. Pytanie jest zadane w sposób narzucający odpowiedź, np. zaznaczając procentową wartość od 0 do 100 - w jakim stopniu zgadzamy się z podanym stwierdzeniem. Wówczas jednak ciężko rozcztać odpowiedź nie wiem, skoro wszystkie kolory sumują się do stu. Ponadto tak naprawdę nie wiadomo, czym jest ostatni (na samym dole każdego słupka) kolor i co on oznacza. Jedyną sensowną interpretacją ze względu na kolory jest rozpatrzenie trzech odpowiedzi: "Tak", "Nie wiem", "Nie" na pytanie czy zgadzamy się z danym stwierdzeniem - i taki wykres spróbuję stworzyć.

## Poprawa wykresu

```python
labels = ['Bardzo cenię sobie\nniezależność i podążanie\nwłasną ścieżką zawodową',
          'Kobiety i mężczyźni\nróżnią się\npredyspozycjami\ni dlatego rożnie\nsprawdzają się\nw różnych zawodach',
          'Uważam, że kobiety\nsprawdzają się\nw STEM czyli\nw zawodach i naukach\nścisłych, technologii\ninżynierii\
 i matematyce\ntak samo dobrze\njak mężczyźni',
          'Kobiecie jest trudniej\nniż mężczyźnie\nbudować karierę zawodową\nze względu na\nzakładanie rodziny',
          'Jest dla mnie\nbardzo ważne,\naby spełniać oczekiwania\nmoich bliskich',
          'Kobiety wybierają\nkarierę, która mniej\ningeruje w ich\nżycie prywatne',
          'Pociągają mnie zajęcia\nniszowe, niepopularne']
women_yes = [76.8, 51.9, 59.5, 59.4, 48.9, 47.4, 41.6]
women_idk = [19.9, 32.4, 31.4, 26.3, 35.6, 34.7, 39.3]
women_no = [3.3, 15.7, 9, 14.3, 15.5, 17.9, 19.2]
men_yes = [67.3, 56.9, 49.4, 49.1, 55.5, 37.4, 37.3]
men_idk = [28, 33.8, 37.2, 34.3, 34.1, 50.3, 46.9]
men_no = [4.8, 9.3, 13.4, 16.6, 10.4, 12.3, 15.7]

x = np.arange(len(labels))
width = 0.36

fig, ax = plt.subplots(figsize=(19, 10))
rects1 = ax.bar(x - 5*width/6, women_yes, width/3, label='Kobiety, "TaK"', zorder=2)
rects2 = ax.bar(x - width/2, women_idk, width/3, label='Kobiety, "Nie wiem"', zorder=2)
rects3 = ax.bar(x - width/6, women_no, width/3, label='Kobiety, "Nie"', zorder=2)
rects4 = ax.bar(x + width/6, men_yes, width/3, label='Mężczyźni, "Tak"', zorder=2)
rects5 = ax.bar(x + width/2, men_idk, width/3, label='Mężczyźni, "Nie wiem"', zorder=2)
rects6 = ax.bar(x + 5*width/6, men_no, width/3, label='Mężczyźni, "Nie"', zorder=2)


ax.set_ylim([0,100])
ax.set_title('\nCzy zgadzasz się ze stwierdzeniem...\n', fontsize=25)
ax.set_yticks([0, 20, 40, 60, 80, 100])
ax.set_yticklabels(('0%', '20%', '40%', '60%', '80%', '100%')) 
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend(prop={'size': 16})
for i in range(20, 101, 20):
    plt.axhline(i, color='black', linestyle='-', linewidth=0.4, zorder=1)
for i in range(10, 91, 20):
    plt.axhline(i, color='black', linestyle='--', linewidth=0.1, zorder=1)
plt.show()
```

## Analiza

Przede wszystkim, w tym momencie jasne jest co przedstawia wykres - nie trzeba się kilka minut zastanawiać, co autor chciał przedstawić.\
Co zostało poprawione:
* zostały wyodrębnione słupki każdy osobno, pozwala to na lepsze porównywanie interesujących nas wartości (wcześniej łatwo było porównywać ilość odpowiedzi "Tak" u mężczyzn i kobiet, ale już o wiele trudniej "Nie wiem";
* nie ma już zlewających się kolorów (tak jak w wykresie oryginalnym zlewała się odpowiedź "nie wiem" z odpowiedzią, która nie posiadała nawet swojego oznaczenia na legendzie;
* dodana czytelniejsza legenda.
* dodane poziome linie pomocniczne, wobec czego nie podpisywałem już dokładnymi wartościami słupków - można je oszacować z wykresu.
