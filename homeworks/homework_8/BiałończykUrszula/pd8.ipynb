## Poprawiony wykres

```python
przychody = [109.7, 51, 41.2, 30.1, 25.9, 22.8, 20.5, 20.5, 18.1, 17.8]
przychody_labels = ["PKN Orlen", "Jeronimo Martins Polska", "PGNiG", "Grupa Lotos", "PGE", "Eurocash", "KGHM",
                   "Orlen Paliwa", "Tauron", "Volkswagen Poznań"]

wzrost = [21, 20, 18, 14, 14, 14, 14, 11, 11, 11, 11]
wzrost_labels = ["budownictwo", "wytwarzanie energii", "usługi", "informatyka", "prod. paliw", "sprzedaż detaliczna",
                "transport", "media i wydawnictwa", "prod. aut, przyczep i statków", "prod. maszyn i urządzeń", "prod. metali"]
```

```python
plt.bar(przychody_labels, przychody)
plt.xticks(przychody_labels, rotation='vertical')
plt.ylabel('Przychody w mld zł.')
plt.xlabel('Firma')
plt.title('Największe firmy w Polsce (przychody w mld zł, bez sektora finansowego)', loc='center')

for index,data in enumerate(przychody):
    plt.text(x=index-0.2 , y =data+1 , s=f"{data}" , fontdict=dict(fontsize=8))
plt.show()
```

```python
plt.bar(wzrost_labels, wzrost)
plt.xticks(wzrost_labels, rotation='vertical')
plt.ylabel('Średnia w procentach')
plt.xlabel('Branża')
plt.title('Wzrost przychodów w branżach (średnio w proc. r./r.)', loc='center')

for index,data in enumerate(wzrost):
    plt.text(x=index-0.2 , y =data+0.2 , s=f"{data}" , fontdict=dict(fontsize=8))
plt.show()
```
