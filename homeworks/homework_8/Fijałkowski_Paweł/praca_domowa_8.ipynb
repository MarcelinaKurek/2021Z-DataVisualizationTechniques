## Poprawienie wykresu - matplotlib

W poniższej pracy poprawię wykres obrazujący stosunek społeczeństwa do nowo wyprodukowanych szczepionek przeciwko COVID-19.

Wykres znajduje się na stronie Centrum Badań Opinii Społecznych(CBOS).

[Link do artykułu.](https://www.cbos.pl/PL/publikacje/news/2020/41/newsletter.php)

Z racji na spore podziały społeczne w wyżej wymienionej kwestii, poszczególne odpowiedzi są porównywalnie często wybierane przez respondentów, tak więc trudno je przanalizować na wykresie kołowym (tak jak zostało to niefortunnie zrobione w oryginale).

Uważam, że dużo czytelniejszy rezultat możemy otrzymać poprzez przedstawienie danych na wykresie słupkowym.

```python
import matplotlib.pyplot as plt
```

```python
fig, ax = plt.subplots()
answers = ["Zdecydowanie tak","Raczej tak","Raczej nie","Zdecydowanie nie","Trudno powiedzieć"]
values = [16,20,20,27,17]
ax.barh(answers,values)
ax.set_xlabel("Procent respondentów")
```

Twierdzę, że poprawiony wykres jest znacznie czytelniejszy, a kolejność przedstawienia zmiennych od najbardziej radykalnych przeciw szczepieniom (Zdecydowanie nie) do pełnej aprobaty (Zdecydowanie tak) pozwala ujrzeć trendy w myśleniu Polaków.
