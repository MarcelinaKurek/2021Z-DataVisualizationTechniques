# Poprawiony

```python
value = [1800,1000, 100, 100, 60, 1]
names = ["Motorbike", "Automobile", "Truck", "Cessna","Motor boat",  "Boeing 247"]
```

```python
plt.figure(figsize=(15,7.5))
ax = plt.axes()        
ax.yaxis.grid()#linie równoległe
ax.set_axisbelow(True)#schowanie lini pod słupkami
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
sns.barplot(names, value, color = "#9ecae1", edgecolor='black')

plt.title('How far can you go with One Barrel Oil?', fontsize=15)
plt.xticks(fontsize=13)
plt.yticks(fontsize=13)
plt.text(-0.65,1850, "km", fontsize=12)
plt.text(-0.1,1810, "1800", fontsize=12)
plt.text(0.9,1010, "1000", fontsize=12)
plt.text(1.9,110, "100", fontsize=12)
plt.text(2.9,110, "100", fontsize=12)
plt.text(3.95,70, "60", fontsize=12)
plt.text(4.97,11, "1", fontsize=12)
```

# Podsumowanie

Na poprawionym wykresie słupki są ustawione w kolejnosći malejącej oraz wartości są przedstawiane jako długość słupka, gdzie w wykresie oryginalnym są one przedstawniane jako powierzchnia słupków. Obie te zmiany sprawiły, że łatwiej jest nam porównywać wartości.
