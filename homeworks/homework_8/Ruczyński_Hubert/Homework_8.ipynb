## Wykres poprawiony

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
data={'Year' : ("2005","2006","2007","2008","2009","2010","2011","2012","2013","2014","2015","2016","2017","2018","2019"), 
      'Deforestation' : (19000,14500,11500,13000,7500,7000,6500,4500,6000,5000,6500,8000,7000,7500,11000)}
df=pd.DataFrame(data)
sns.set(rc={'figure.figsize':(14,10)})
sns.set_style("whitegrid")
```

```python
colors=[]
for i in range (0,14):
    if(df.Deforestation[i]>=10000):
        colors.append('tan')
    else:
        colors.append('lightgreen')
sns.barplot(data = df, x = "Year", y="Deforestation", palette=colors)
plt.title("Estimated deforestation in the Brazilian Amazon", size=30)
plt.xlabel("Year", size=17)
plt.ylabel("Deforestation (km^2)", size=17)
plt.show()
```

## Uzasadnienie zmian

Zauważyłem jednak konwencję kolorowania autora (to, że wyższe słupki przechodziły w kolor brązowy) i postanowiłem ją poprawić.
