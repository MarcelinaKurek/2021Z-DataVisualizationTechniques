# Poprawiona wizualizacja

```python
val_year = []
val_month = []
months = ["2010 I", "2010 VI", "2010 XI", "2011 IV", "2011 IX", "2012 II", " 2012 VII", "2012 XII", "2013 V", "2013 X", "2014 III", "2014 VIII", "2015 I", "2015 VI", "2015 XI", "2016 IV", "2016 IX", "2017 VII", "2017 XII", "2018 V", "2018 X", "2019 III", "2019 VIII", "2020 I", "2020 VI", "2020 XI"]

#sam robię wartości...
import random

for i in range(0,26):
    n = random.random() + 3.5
    val_year.append(n)


for i in range(0,26):
    n = random.random() - 0.5
    val_month.append(n)


plt.xticks(rotation=90)
plt.grid(axis='y')
p1,  = plt.plot(months, val_year, label='rok do roku')


p2, = plt.plot(months, val_month, label='miesiąc do miesiąca')
plt.legend(handles=[p1, p2], bbox_to_anchor=(1.05, 1), loc='upper left')
plt.title("Zmiany procentowe wskaźnika cen towarów i usług konsumpcyjnych (CPI)")
plt.ylabel("Zmiana w %")
plt.show()
```

# Omówienie

Pierwszą poprawą jest to, że są to nałożone linie na ten sam wykres, a nie, tak jak w oryginale, dwa wykresy złączone razem. Drugą poprawą, jest przesunięcie legendy poza wykres.
