Zdecydowałem się poprawić wizualizację porównującą występy trzech środkowych reprezentantów Polski. Uważam, że obecna forma, czyli tabela, słabo pozwala porównać między sobą poszczególne statystyki. Dodatkowo niektóre pola zostały przedstawione procentowo, a inne nie. W mojej poprawionej wersji wszystkie statystyki zostały przedstawione procentowo (w przypadku niektórych statystyk za 100% została przyjęta maksymalna wartość).

Oryginał znajduje się tutaj: https://youtu.be/rlx7yPENE7U?t=2157

```python
import pandas
import matplotlib.pyplot as plt
import numpy as np

maksymalne = [90/100,52/100,100/100,1/100,1/100,100/100,100/100,9/100]

df = pandas.DataFrame(dict(
    graph=['Minuty gry','Kontakty z piłką','Celne podania','Kluczowe podania','Strzały na bramkę','Wygrane pojedynki',
         'Długie piłki','Tracił posiadanie'],
    krychowiak = [90,52,96,0,0,5,80,2],
    moder = [45,33,67,0,0,0,50,9],
    linetty = [74,32,81,0,0,38,100,9],
    krychowiak2 = [90/maksymalne[0],52/maksymalne[1],96/maksymalne[2],
               0/maksymalne[3],0/maksymalne[4],25/maksymalne[5],80/maksymalne[6],2/maksymalne[7]],
    moder2 = [45/maksymalne[0],33/maksymalne[1],67/maksymalne[2],
             0/maksymalne[3],0/maksymalne[4],0/maksymalne[5],50/maksymalne[6],9/maksymalne[7]],
    linetty2 = [74/maksymalne[0],32/maksymalne[1],81/maksymalne[2],
             0/maksymalne[3],0/maksymalne[4],38/maksymalne[5],100/maksymalne[6],9/maksymalne[7]]))

ind = np.arange(len(df))
width = 0.3

fig, ax = plt.subplots()
ax.barh(ind + 2*width, df.krychowiak2, width, color='#1B9E77', label='Krychowiak')
ax.barh(ind+width, df.moder2, width, color='#D95F02', label='Moder')
ax.barh(ind, df.linetty2, width, color='#7570B3', label='Linetty')
        
ax.set(yticks=ind + width, yticklabels=df.graph, ylim=[2*width - 1, len(df)])
ax.legend()
ax.set_title("Porównanie występów środkowych pomocników")
ax.invert_yaxis()
plt.show()
```
