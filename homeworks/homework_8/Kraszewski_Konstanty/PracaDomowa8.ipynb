## Poprawiony wykres

```python
source = ['Nuclear', 'Gas', 'Wind', 'Hydro', 'Brown coal', 'Black coal', 'Biomass', 'Solar', 'Other fossil']
percentage = [25.5, 21.6, 13.4, 10.8, 7.8, 6.7, 6.2, 4.2, 3.6]

fig, ax = plt.subplots(figsize = (12,9))

rects = ax.bar(source, percentage)

plt.xlabel('energy source', size = 15)
plt.ylabel('Part of all energy in pc.', size = 15)
plt.title("EU-27 Electrical Energy Sources in 2019", size = 23)


for rect in rects:
        height = rect.get_height()
        ax.text(rect.get_x() + rect.get_width()/2., height,
                '%s' % str(height)+'%',
                ha='center', va='bottom')
        
plt.show()
```

Poprawiony wykres eliminuje niepotrzebne kolory i zamienia wykres na słupkowy do łatwiejszego porównywania danych (prostokąty z pierwszego wykresu jeszcze trudniej porównywać niż kąty na wykresie kołowym.
