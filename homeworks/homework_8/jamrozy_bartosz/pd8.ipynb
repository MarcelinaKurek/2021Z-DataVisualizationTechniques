## Poprawka
Podwójny wykres pączkowy nie jest za bardzo czytelny.

```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
sns.set(rc={'figure.figsize':(20,8.27)})
sns.set_style("whitegrid")

def numerate_bar(s):
    for idx, ax in np.ndenumerate(s):
        for p in ax.patches:
            _x = p.get_x() + p.get_width() / 2 
            _y = p.get_y() + p.get_height() + 20
            value = int(p.get_height())
            ax.text(_x, _y, value, ha="center")
            
liczba_turbin=[16,42,121,323]
moc=[101,252,1012,2261]
company=["SGRE","MHI Vestas","  GE Renewable Energy","Senvion"]

fig, ax =plt.subplots(1,2)

s=sns.barplot(x=company,y=moc,ax=ax[0])
numerate_bar(s)            
s.set_title("Przyłączona moc w MW")

p=sns.barplot(x=company,y=liczba_turbin,ax=ax[1])
numerate_bar(p)            
p.set_title("Liczba dostarczonych turbin")

plt.show()
```
