### Poprawiony wykres w Pythonie

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
countries =['USA', 'Wielka Brytania', 'Niemcy', 'Hiszpania', 'Wlochy', 'Australia', 'Francja', 'Polska', 'Portugalia', 'Rumunia', 'Austria', 'Czechy', 'Holandia', 'Szwajcaria', 'Slowacja', 'Belgia', 'Izrael', 'Szwecja']
total_tests_per_country = [115347287, 26679020, 18129900, 13689776, 12197500, 7935463, 11900095, 3599722, 2764867, 2627544, 1756856, 1515912, 2656731, 1470894, 528367, 3545600, 426328, 1789328]
tests_per_million =[347922, 392435, 216200, 292768, 201821, 310224, 182201, 95483, 271371, 136837, 194765, 141483, 154954, 169617, 96764, 305393, 426328, 176874]
```

```python
df=pd.DataFrame(countries)
df['total']=total_tests_per_country
df['milion']=tests_per_million
df=df.sort_values(by='milion',ascending=False)
df=df.rename(columns={0:'country'})
dt=df
dt
```

```python
fig, ax = plt.subplots(2,1,figsize=(8, 12))

for i in range(len(df['total'])):
    dt['total'][i]=df['total'][i]/1000000
    dt['milion'][i]=df['milion'][i]/1000
wart
ax[1].barh(dt['country'], dt['total'], color=["#6699ff"], height=0.8, label="Liczba wszystkich testów w milionach")
ax[1].legend(loc='upper center')
ax[1].set_ylabel("Państwo", fontsize=12)
ax[0].barh(dt['country'], dt['milion'], color=["#6699ff"], height=0.8, label="Liczba tysięcy testów wykonanych \nna milion mieszkańców")
ax[0].legend(loc='upper center')
ax[0].set_ylabel("Państwo", fontsize=12)

fig.suptitle("Liczba testów wykonanych w poszczególnych krajach na świecie", fontsize=15)
plt.show()
```
